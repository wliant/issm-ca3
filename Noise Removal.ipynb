```python
classes = {
            "walk":0,
            "bike":1,
            "bus":2,
            "taxi": 3,
            "car":3, 
            "subway":4, 
            "train":4, 
            "airplane":6, 
            "boat":7, 
            "run":8, 
            "motorcycle":10
        }
```

```python
df = df.replace({'label':classes})
df['label'].unique()
```

```python
df['label']=df['label'].astype(int)
```

```python
selected_classes = df[df['label']<5]
selected_classes['label'].unique()
```

```python
inv_map = {v: k for k, v in classes.items()}
```

```python
inv_map
```

```python
selected_classes = selected_classes.replace({'label':inv_map})
selected_classes['label'].unique()
```

```python
mode_mean=selected_classes.groupby(['label'])['speed_mean'].mean().to_frame(name = 'mode_mean').reset_index()
mode_mean.head()
```

```python
data_mean= selected_classes.merge(mode_mean, left_on='label', right_on='label')
data_mean.head()
```

```python
mode_std=selected_classes.groupby(['label'])['speed_mean'].std().to_frame(name = 'mode_std').reset_index()
mode_std.head()
```

```python
stdev= data_mean.merge(mode_std, left_on='label', right_on='label')
stdev.head()
```

```python
def std_dev(speed_mean,mode_mean,mode_std): 
    if (speed_mean < mode_mean - 2 * mode_std):
        return 1
    elif (speed_mean > mode_mean + 2 * mode_std):
        return 1
    else:
        return 0
```

```python
stdev["to_remove"] = stdev.apply(lambda x: std_dev(x["speed_mean"],x["mode_mean"],x["mode_std"]), axis=1)
stdev.head()
```

```python
len(stdev)
```

```python
noise_removed = stdev[stdev['to_remove']!=1]
len(noise_removed)
```
